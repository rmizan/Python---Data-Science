### Problem 5
For each state, what is the percentage of fatal accidents involved at least one drunk driver? If the column, DRUNK_DR, has a 0, then no drunk drivers were involved. Any number larger than 0 indicates the number of drunk drivers involved in the accident. 

```python
df=pd.DataFrame(columns=['STATE','COUNT'])
for i in table.keys():
    df_i= pd.read_csv("C:/Users/mrahman1s/Documents/CSCI5015/Week 8/FARS/FARS/accident_{:02d}.csv".format(i))
    a= df_i['DRUNK_DR'].value_counts()
    drunk_accidents= (len(df_i)-a[0])/len(df_i)
    df1=pd.DataFrame({'STATE':[table[i]],'COUNT':[drunk_accidents*100]})
    df=df.append(df1,ignore_index=True)
    #print("The % of accidents for state {} involving drunk drivers is {:.4f}".format(table[i],drunk_accidents*100))
df
```

### Problem 8
For each state, what is its fatal accident rate per 10,000 people? To calculate this, count the number of accidents in a state, divide by the state's 2016 population estimate from the nst-est2017-alldata.csv Census data file, and then multiply by 10000. Output the states' rates in order from highest to lowest. 

```python
rate={}
df=pd.DataFrame(columns=['STATE','COUNT'])

for i in table.keys():
    df_i= pd.read_csv("C:/Users/mrahman1s/Documents/CSCI5015/Week 8/FARS/FARS/accident_{:02d}.csv".format(i))
    accidents= len(df_i)
    index=df_all.index[df_all['NAME'] == table[i]].tolist()[0]
    population= df_all['POPESTIMATE2016'][index]
    rate[i]= (accidents/population)*10000
rate_sorted=sorted(rate.items() , reverse=True, key=lambda x: x[1])
for i in rate_sorted:
    df1=pd.DataFrame({'STATE':[table[i[0]]],'COUNT':[i[1]]})
    df=df.append(df1,ignore_index=True)
    #print("The rate of fatalaccidents of state {} is {}".format(table[i[0]],i[1]))
df
```
