Problem 2: How many unique weather station ids (STATION fields) are in the data file? Provide the count and then list them.

```python
#2)unique station ids
ids=[]
for row in data:
    if row['STATION'] not in ids:
        ids.append(row['STATION'])
```

Problem 3: List the unique station ids (STATION field) next to their name (NAME field)

Problem 4: List date range (DATE) for each station (STATION). Stations come and go, so their data is only available for limited date ranges. For each station, list the date of the earliest record in the file and the date of the last record in the file. 

```python
#4)listing date range
for i in ids:
    min_date=None
    max_date=None
    for row in data:
        if i == row['STATION']:
            d=time.strptime(row['DATE'],"%m/%d/%Y")
            if min_date==None:
                min_date=d
            elif min_date>d:
                min_date=d
            if max_date==None:
                max_date=d
            elif max_date<d:
                max_date=d
    print(i," : Start date :",min_date.tm_mon,"/",min_date.tm_mday,"/",min_date.tm_year," End date :",max_date.tm_mon,"/",max_date.tm_mday,"/",max_date.tm_year)
```

Problem 5: List the station id with the highest elevation and the station with the lowest elevation. 

```python
#5)
min_elevation=None
max_elevation=None
min_id=None
max_id=None
for i in ids:
    for row in data:
        if i==row['STATION']:
            if min_elevation==None:
                min_elevation=float(row['ELEVATION'])
                min_id=i
            elif min_elevation>float(row['ELEVATION']):
                min_elevation=float(row['ELEVATION'])
                min_id=i
            if max_elevation==None:
                max_elevation=float(row['ELEVATION'])
                max_id=i
            elif max_elevation<float(row['ELEVATION']):
                max_elevation=float(row['ELEVATION'])
                max_id=i
print("Minimun/lowest Elevation is: ",min_elevation," of id :",min_id)
print("MAximum/highest Elevation is: ",max_elevation," of id :",max_id)
```

```python
#Common code for questions 6,7,8,9,10)
max_tmax=None
min_tmax=None
max_tmin=None
min_tmin=None
prcp=None
for row in data:
    if row['TMAX']!=None:
        if min_tmax==None:
            min_tmax=row['TMAX']
        elif min_tmax>row['TMAX']:
            min_tmax=row['TMAX']
        if max_tmax==None:
            max_tmax=row['TMAX']
        elif max_tmax<row['TMAX']:
            max_tmax=row['TMAX']
    if row['TMIN'] !=None:
        if min_tmin==None:
            min_tmin=row['TMIN']
        elif min_tmin>row['TMIN']:
            min_tmin=row['TMIN']
        if max_tmin==None:
            max_tmin=row['TMIN']
        elif max_tmin<row['TMIN']:
            max_tmin=row['TMIN']
    if row['PRCP']!=None:
        if prcp==None:
            prcp=row['PRCP']
        elif prcp<row['PRCP']:
            prcp=row['PRCP']
```
